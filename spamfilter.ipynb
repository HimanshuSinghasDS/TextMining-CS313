```python
import re
```

```python
def lower_cases(text):
 return text.lower()

def remove_char(text):
 pattern = r'[^a-zA-Z0-9\s]'
 text = re.sub(pattern, '', text)
 return text
```

```python
def remove_brackets(text):
 return re.sub('\[[^]]*\]', '', text)
```

```python
def remove_noise(text):
 text = remove_brackets(text)
 text = lower_cases(text)
 text = remove_char(text)
 return text
```

```python
df['Message'] = df['Message'].apply(remove_noise)
```

```python
df.head()
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
def stem_words(text):
 ps = PorterStemmer()
 stem_list = [ps.stem(word) for word in text.split()]
 text = ''.join(ps.stem(word) for word in text)

 return text
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
stoplis = stopwords.words('english')
```

```python
def rm_stp_w(text):
  xtext = []
  for word in text.split():
    if word not in stoplis:
      xtext.append(word)

  final = xtext[:]
  xtext.clear()
  return final
```

```python
df['Message'] = df['Message'].apply(rm_stp_w)
```

```python
df['Message']
```

```python
def join_back(text):
  return ' '.join(text)
```

```python
df['Message'] = df['Message'].apply(join_back)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```
