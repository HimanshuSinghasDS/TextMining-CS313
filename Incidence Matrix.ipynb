```python
import os
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def cbtdm(directory):
    vectorizer = CountVectorizer(input='filename', binary=True)

    X = vectorizer.fit_transform([os.path.join(directory, file) for file in os.listdir(directory) if file.endswith('.txt')])

    words = vectorizer.get_feature_names_out()

    binary_matrix = X.toarray()

    return words, binary_matrix.transpose()
    
```

```python
def gtfn(directory):
    files = []
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            files.append(filename)
    return files
```

```python
directory = 'TM'
filenames = gtfn(directory)
print(filenames)
```

```python
words, binary_matrix = cbtdm(directory)
```

```python
import pandas as pd
X = pd.DataFrame(binary_matrix, index = words, columns = filenames)
```
