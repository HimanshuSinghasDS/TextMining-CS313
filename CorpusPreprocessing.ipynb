```python
#importing essential libraries
import nltk
nltk.download('punkt') #download resources for tokenization & puntuations

#for reading the text file
file = open("/content/end.txt" , "r")
text = file.read()

# Sentence Tokenization
from nltk.tokenize import sent_tokenize

token_sent = sent_tokenize(text)
print(token_sent)

# Word Tokenization
from nltk.tokenize import word_tokenize

token_word = word_tokenize(text)
print(token_word)
```

```python
import string

#punctuations
punctuations = list(string.punctuation)

token = []

for i in token_word:
  if i not in punctuations:
      token.append(i)

print(token)
```

```python
#lowercasing
ltoken = []
for token in token_word:
  ltoken.append(token.lower())
print(ltoken)

#truecasing
import nltk
#!pip install truecase
from truecase import get_true_case

#truecasing(Sentences)
true_sent = []
for sent in token_sent:
    true_sent.append(get_true_case(sent))
print(true_sent)

#truecasing(Words)
true_words = []
for words in true_sent:
    true_words.extend(word_tokenize(words))
print(true_words)

#creating dataframe
import pandas as pd
d = {'Original' : token_word, 'Lowercase' : ltoken, 'Truecase' : true_words}
df = pd.DataFrame(d)
df
```
