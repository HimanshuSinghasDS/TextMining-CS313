```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

#Stemming
swords = []
for i in ftoken:
    swords.append(ps.stem(i))

print("Stem Words:")
print(swords)
```

```python
#Lemmatization
from nltk.stem.wordnet import WordNetLemmatizer
lem = WordNetLemmatizer()

lwords = []

for i in ftoken:
    lwords.append(lem.lemmatize(i, "v"))

print("Lemmatize Words:")
print(lwords)
```
