```python
from sklearn.datasets import fetch_20newsgroups
```

```python
cats = ['alt.atheism', 'sci.space'] # more categories are also available like 'talk.religion.misc', 'comp.graphics'.
newsgroups_train = fetch_20newsgroups(subset='train', categories=cats)
list(newsgroups_train.target_names)
```

```python
newsgroups_train.filenames.shape
newsgroups_train.target.shape
newsgroups_train.target[:10]
```

```python
newsgroups_train.target.shape
```

```python
newsgroups_train.filenames.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(newsgroups_train.data)
```

```python
vectors = vectors.toarray()
```

```python
vectors
```

```python
import numpy as np
np.unique(newsgroups_train.target)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
newsgroups_test = fetch_20newsgroups(subset='test',categories=cats)
vectors_test = vectorizer.transform(newsgroups_test.data)
```

```python
newsgroups_test.filenames.shape
```

```python
clf = MultinomialNB(alpha=.01)
clf.fit(vectors, newsgroups_train.target)
```

```python
pred = clf.predict(vectors_test)
print(metrics.classification_report(newsgroups_test.target, pred))#average='macro'
```

```python
clf2 = NearestCentroid()
clf2.fit(vectors, newsgroups_train.target)
```
